```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
N = 100000
a = 0.95
X = np.random.normal(0, 7, N)
```

```python
def plot(x, y, legend, ylim=(0,150)):
    fig = plt.figure(figsize=(15, 10))
    plt.xlabel('N', fontsize=15)
    plt.ylabel('Confidence interval lenght', fontsize=15)
    plt.plot(x, y, label=legend)
    plt.ylim(ylim) 
    plt.show()
```

```python
abscissa = np.arange(1,N+1)
sumX2 = np.cumsum(X**2)
left_q = scipy.stats.chi2.ppf((1-a)/2, abscissa)
right_q = scipy.stats.chi2.ppf((1+a)/2, abscissa)
lenghts1 = sumX2 / left_q - sumX2 / right_q

meanX = np.cumsum(X) / abscissa
left_z = abscissa * scipy.stats.norm.ppf((3-a)/4, 0, 1)**2
right_z = abscissa * scipy.stats.norm.ppf((3+a)/4, 0, 1)**2
lenghts2 = N*meanX**2 / left_z - N*meanX**2 / right_z
```

```python
plot(abscissa, lenghts1, 'First Statistic')
```

```python
plot(abscissa, lenghts2, 'Second Statistic', ylim=(0,1e4))
```
